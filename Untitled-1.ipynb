```python
# %%
# ==============================================================================
# PIPELINE ISTITUZIONALE DI REPLICA: OLS STATICA vs REGULARIZED FACTOR KALMAN
# Architecture: SVD/PCA Factor Space, O-U Mean Reversion, EWMA Noise & Tikhonov L2
# ==============================================================================

import warnings
import matplotlib.pyplot as plt
import numpy as np
import optuna
import pandas as pd
from scipy.linalg import svd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import TimeSeriesSplit

# Configurazione ambiente e soppressione log selettiva
plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["font.size"] = 11
warnings.filterwarnings("ignore", category=RuntimeWarning)
optuna.logging.set_verbosity(optuna.logging.WARNING)

# %%
# ==============================================================================
# MODULO 1: INGESTIONE DATI & COSTRUZIONE PANIERI (SENZA TITOLI CORROTTI)
# ==============================================================================
percorso_file_dati = "Dataset3_PortfolioReplicaStrategyErrataCorrige.xlsx"
percorso_file_meta = "Dataset3_PortfolioReplicaStrategy.xlsx"

ticker_bloomberg = (
    pd.read_excel(percorso_file_meta, header=None, skiprows=5, nrows=1)
    .iloc[0]
    .tolist()[1:]
)

dati_grezzi = pd.read_excel(percorso_file_dati, header=None, skiprows=1)
dati_grezzi.columns = ["Data"] + ticker_bloomberg
dati_grezzi["Data"] = pd.to_datetime(dati_grezzi["Data"], format="%d/%m/%Y")
dati_prezzi = dati_grezzi.set_index("Data").sort_index()

# Rimozione del contratto corrotto LLL1 Comdty
if "LLL1 Comdty" in dati_prezzi.columns:
    dati_prezzi = dati_prezzi.drop(columns=["LLL1 Comdty"])

rendimenti = dati_prezzi.pct_change().dropna()

# Benchmark Target Composito Multi-Asset
componenti_indice_target = {
    "HFRXGL Index": 0.50,  # Hedge Fund Research Global Index
    "MXWO Index": 0.25,  # MSCI World Equity Index
    "LEGATRUU Index": 0.25,  # Bloomberg Global Aggregate Bond Index
}

rendimento_target = pd.Series(
    0.0, index=rendimenti.index, name="Indice_Target"
)
for componente, peso in componenti_indice_target.items():
    rendimento_target += rendimenti[componente] * peso

contratti_future = [
    "RX1 Comdty",
    "TY1 Comdty",
    "GC1 Comdty",
    "CO1 Comdty",
    "ES1 Comdty",
    "VG1 Comdty",
    "NQ1 Comdty",
    "TP1 Comdty",
    "DU1 Comdty",
    "TU2 Comdty",
]

date_comuni = rendimento_target.index.intersection(rendimenti.index)
Y = rendimento_target.loc[date_comuni]
C = rendimenti.loc[date_comuni, contratti_future]

print("--- STATO INGESTIONE DATI ---")
print(
    f"Osservazioni totali: {len(Y)} settimane | Universo predittori: {C.shape[1]}"
    " future\n"
)

# %%
# ==============================================================================
# MODULO 2: PARTIZIONAMENTO TEMPORALE RIGOROSO (60% TRAIN / 20% VAL / 20% TEST)
# ==============================================================================
T_totale = len(Y)
idx_train = int(T_totale * 0.60)
idx_val = int(T_totale * 0.80)

Y_train, C_train = Y.iloc[:idx_train], C.iloc[:idx_train]
Y_val, C_val = Y.iloc[idx_train:idx_val], C.iloc[idx_train:idx_val]
Y_test, C_test = Y.iloc[idx_val:], C.iloc[idx_val:]

print("--- PARTIZIONAMENTO CRONOLOGICO ---")
print(
    f"Train Set: {len(Y_train)} obs | Val Set (OOS 1): {len(Y_val)} obs | Test"
    f" Set (OOS 2): {len(Y_test)} obs\n"
)

# %%
# ==============================================================================
# MODULO 3: BASELINE OLS STATICA & ESTRAZIONE FATTORIALE SVD/PCA NON CENTRATA
# ==============================================================================
# 1. Baseline OLS Statica su 10 future
modello_ols = LinearRegression(fit_intercept=False).fit(
    C_train.values, Y_train.values
)
pesi_ols_statici = pd.Series(
    modello_ols.coef_, index=contratti_future, name="Peso_OLS"
)
rendimenti_ols_totali = pd.Series(
    modello_ols.predict(C.values), index=C.index, name="Replica_OLS"
)

# 2. Estrazione Fattori Ortogonali via SVD (Modifica 4) - Strictly In-Sample
# Utilizziamo SVD non centrata per preservare l'assenza di intercetta nei rendimenti
k_fattori = 4
U, s, Vt = svd(C_train.values, full_matrices=False)
V_k = Vt[:k_fattori, :].T  # Matrice di proiezione ortogonale: Shape (N, k)

# Proiezione dell'intero dataset nello spazio dei fattori ortogonali F_t = C_t @ V_k
F_totale = pd.DataFrame(
    C.values @ V_k,
    index=C.index,
    columns=[f"Fattore_{i+1}" for i in range(k_fattori)],
)
F_train = F_totale.iloc[:idx_train]

# Calcolo Beta OLS Target di lungo periodo nello spazio dei fattori
modello_ols_fattori = LinearRegression(fit_intercept=False).fit(
    F_train.values, Y_train.values
)
beta_target_ols = modello_ols_fattori.coef_

var_spiegata = np.sum(s[:k_fattori] ** 2) / np.sum(s**2) * 100
print("--- ESTRAZIONE FATTORI ORTOGONALI (SVD/PCA) ---")
print(
    f"Numero fattori estratti: {k_fattori} | Varianza spiegata in-sample:"
    f" {var_spiegata:.2f}%"
)
print(f"Norma beta target di equilibrio: {np.linalg.norm(beta_target_ols):.4f}\n")

# %%
# ==============================================================================
# MODULO 4: FILTRO DI KALMAN REGOLARIZZATO A FATTORI (RFKF)
# ==============================================================================
class RegularizedFactorKalmanFilter:
    """Filtro di Kalman con transizione di Ornstein-Uhlenbeck, rumore di osservazione

    adattivo EWMA e regolarizzazione bayesiana di Tikhonov (L2 penalty) sullo stato.
    """

    def __init__(self, phi, q_std, gamma_ewma, sigma_reg, beta_target, V_proj):
        self.phi = phi  # Persistenza O-U (Mean Reversion)
        self.q_std = q_std  # Deviazione standard rumore di processo
        self.gamma = gamma_ewma  # Moltiplicatore varianza EWMA per rumore osservazione
        self.sigma_reg = sigma_reg  # Dev std penalizzazione L2 verso beta target
        self.beta_target = beta_target  # Beta di lungo periodo (Eq. fondamentale)
        self.V_proj = V_proj  # Matrice di caricamento asset (N, k)

    def fit_and_predict(self, Y_vals, F_vals, C_vals):
        T, k = F_vals.shape
        N = C_vals.shape[1]

        self.beta_exante = np.zeros((T, k))
        self.pesi_exante = np.zeros((T, N))
        self.y_hat_exante = np.zeros(T)

        # Inizializzazione stato e covarianza a priori
        beta_prev = self.beta_target.copy()
        P_prev = np.eye(k) * 0.01

        # Inizializzazione varianza EWMA sul primo mese
        var_ewma = np.var(Y_vals[:4]) if T >= 4 else 0.0001
        lam_ewma = 0.94

        Q = np.eye(k) * (self.q_std**2)
        R_reg_inv = (
            np.eye(k) * (1.0 / (self.sigma_reg**2)) if self.sigma_reg > 0 else None
        )

        for t in range(T):
            F_t = F_vals[t].reshape(1, -1)
            y_t = Y_vals[t]

            # --- MODIFICA 2: RUMORE DI OSSERVAZIONE ADATTIVO EWMA ---
            if t > 0:
                var_ewma = lam_ewma * var_ewma + (1.0 - lam_ewma) * (
                    Y_vals[t - 1] ** 2
                )
            R_t = max(self.gamma * var_ewma, 1e-6)

            # --- MODIFICA 1: TRANSIZIONE MEAN-REVERTING DI ORNSTEIN-UHLENBECK ---
            # beta_{t|t-1} = phi * beta_{t-1|t-1} + (1 - phi) * beta_target
            beta_pred = (
                self.phi * beta_prev + (1.0 - self.phi) * self.beta_target
            )
            P_pred = (self.phi**2) * P_prev + Q

            y_hat_t = (F_t @ beta_pred).item()

            # Mapping dai beta fattoriali ai pesi dei 10 future: w_{t|t-1} = V_k @ beta_{t|t-1}
            w_pred = self.V_proj @ beta_pred

            self.beta_exante[t] = beta_pred
            self.pesi_exante[t] = w_pred
            self.y_hat_exante[t] = y_hat_t

            # --- STEP 1: AGGIORNAMENTO DI MERCATO (INNOVAZIONE PURA) ---
            S_t = (F_t @ P_pred @ F_t.T).item() + R_t
            K_t = (P_pred @ F_t.T) / S_t
            beta_up1 = beta_pred + K_t.flatten() * (y_t - y_hat_t)

            I_KF = np.eye(k) - K_t @ F_t
            P_up1 = I_KF @ P_pred @ I_KF.T + (K_t @ K_t.T) * R_t

            # --- MODIFICA 3: REGOLARIZZAZIONE DI TIKHONOV (PSEUDO-OSSERVAZIONE L2) ---
            # Restringe analiticamente la norma del beta verso l'equilibrio OLS se l'incertezza sale
            if R_reg_inv is not None:
                S_reg = P_up1 + np.eye(k) * (self.sigma_reg**2)
                K_reg = P_up1 @ np.linalg.inv(S_reg)
                beta_new = beta_up1 + K_reg @ (self.beta_target - beta_up1)
                I_Kreg = np.eye(k) - K_reg
                P_new = I_Kreg @ P_up1
            else:
                beta_new, P_new = beta_up1, P_up1

            beta_prev, P_prev = beta_new, P_new

        return self


# %%
# ==============================================================================
# MODULO 5: META-OTTIMIZZAZIONE WALK-FORWARD NESTED CV SUL TRAIN SET
# ==============================================================================
print(
    "Avvio ottimizzazione Optuna (Walk-Forward CV 4-Folds all'interno del Train Set"
    " 0%-60%)..."
)


def funzione_obiettivo_rfkf(trial):
    # Spazio degli iperparametri architetturali
    phi = trial.suggest_float("phi", 0.80, 0.999)
    q_std = trial.suggest_float("q_std", 1e-5, 0.05, log=True)
    gamma_ewma = trial.suggest_float("gamma_ewma", 0.1, 15.0, log=True)
    sigma_reg = trial.suggest_float("sigma_reg", 1e-4, 0.5, log=True)

    Y_tr = Y_train.values
    F_tr = F_train.values
    C_tr = C_train.values

    cv_temporale = TimeSeriesSplit(
        n_splits=4, test_size=int(len(Y_train) * 0.15)
    )
    errori_tracking_folds = []

    for idx_sub_train, idx_sub_val in cv_temporale.split(Y_tr):
        ind_fine = idx_sub_val[-1] + 1
        modello_cv = RegularizedFactorKalmanFilter(
            phi=phi,
            q_std=q_std,
            gamma_ewma=gamma_ewma,
            sigma_reg=sigma_reg,
            beta_target=beta_target_ols,
            V_proj=V_k,
        )
        modello_cv.fit_and_predict(
            Y_tr[:ind_fine], F_tr[:ind_fine], C_tr[:ind_fine]
        )

        predizioni_val = modello_cv.y_hat_exante[idx_sub_val]
        valori_veri_val = Y_tr[idx_sub_val]
        te_fold = np.std(predizioni_val - valori_veri_val) * np.sqrt(52)
        errori_tracking_folds.append(te_fold)

    return np.mean(errori_tracking_folds)


studio_optuna = optuna.create_study(direction="minimize")
studio_optuna.optimize(funzione_obiettivo_rfkf, n_trials=120)

m_phi = studio_optuna.best_params["phi"]
m_q_std = studio_optuna.best_params["q_std"]
m_gamma = studio_optuna.best_params["gamma_ewma"]
m_sigma_reg = studio_optuna.best_params["sigma_reg"]

print(
    f"Ottimizzazione terminata. TE medio in CV interna:"
    f" {studio_optuna.best_value*100:.2f}%"
)
print(f"Iperparametri Ottimi -> Phi (Mean-Rev): {m_phi:.4f} | Q_std: {m_q_std:.5f}")
print(f"Gamma EWMA (Noise): {m_gamma:.2f} | Sigma Regolarizzazione L2: {m_sigma_reg:.5f}\n")

# --- ESECUZIONE GLOBALE DEL MODELLO RFKF CON IPERPARAMETRI OTTIMI ---
modello_rfkf = RegularizedFactorKalmanFilter(
    phi=m_phi,
    q_std=m_q_std,
    gamma_ewma=m_gamma,
    sigma_reg=m_sigma_reg,
    beta_target=beta_target_ols,
    V_proj=V_k,
)
modello_rfkf.fit_and_predict(Y.values, F_totale.values, C.values)

df_pesi_rfkf = pd.DataFrame(
    modello_rfkf.pesi_exante, index=Y.index, columns=C.columns
)
rendimenti_rfkf_grezzi = pd.Series(modello_rfkf.y_hat_exante, index=Y.index)

# --- CALCOLO ESATTO DEL TURNOVER EX-POST CON DRIFT DEI PESI ---
rendimenti_rfkf_netti = pd.Series(0.0, index=Y.index, name="Replica_RFKF_Netta")
serie_turnover = pd.Series(0.0, index=Y.index, name="Turnover")
costo_transazione_unitario = 0.0005  # 5 bps

for t in range(len(Y)):
    if t == 0:
        rendimenti_rfkf_netti.iloc[t] = rendimenti_rfkf_grezzi.iloc[t]
        serie_turnover.iloc[t] = np.abs(df_pesi_rfkf.iloc[t]).sum()
    else:
        pesi_t_meno_1 = df_pesi_rfkf.iloc[t - 1]
        rendimento_asset_t = C.iloc[t]
        pesi_con_drift = (pesi_t_meno_1 * (1 + rendimento_asset_t)) / (
            1 + rendimenti_rfkf_grezzi.iloc[t]
        )

        turnover_t = np.abs(df_pesi_rfkf.iloc[t] - pesi_con_drift).sum()
        serie_turnover.iloc[t] = turnover_t
        rendimenti_rfkf_netti.iloc[t] = (
            rendimenti_rfkf_grezzi.iloc[t]
            - turnover_t * costo_transazione_unitario
        )

# Segregazione definitiva delle performance
rendimenti_ols_train = rendimenti_ols_totali.iloc[:idx_train]
rendimenti_ols_val = rendimenti_ols_totali.iloc[idx_train:idx_val]
rendimenti_ols_test = rendimenti_ols_totali.iloc[idx_val:]

rendimenti_kf_train = rendimenti_rfkf_netti.iloc[:idx_train]
rendimenti_kf_val = rendimenti_rfkf_netti.iloc[idx_train:idx_val]
rendimenti_kf_test = rendimenti_rfkf_netti.iloc[idx_val:]

pesi_kf_train = df_pesi_rfkf.iloc[:idx_train]
pesi_kf_val = df_pesi_rfkf.iloc[idx_train:idx_val]
pesi_kf_test = df_pesi_rfkf.iloc[idx_val:]

# %%
# ==============================================================================
# MODULO 6: MOTORE DI VALUTAZIONE PERFORMANCE & TABELLE MULTI-INDICE
# ==============================================================================
def calcola_metriche_istituzionali(
    r_target, r_replica, pesi_asset=None, e_target=False, freq=52.0
):
    T = len(r_replica)
    if T == 0:
        return ["-"] * 8

    ricchezza_cumulata = (1.0 + r_replica).cumprod()
    cagr = (ricchezza_cumulata.iloc[-1] / ricchezza_cumulata.iloc[0]) ** (
        freq / T
    ) - 1.0
    volatilita_annua = r_replica.std() * np.sqrt(freq)
    indice_sharpe = cagr / volatilita_annua if volatilita_annua != 0 else 0.0
    max_drawdown = (
        1.0 - ricchezza_cumulata / ricchezza_cumulata.cummax()
    ).max()

    if e_target:
        return [
            f"{cagr*100:.2f}%",
            f"{volatilita_annua*100:.2f}%",
            f"{indice_sharpe:.2f}",
            f"{max_drawdown*100:.2f}%",
            "-",
            "-",
            "1.0000",
            "1.00",
        ]

    rendimenti_attivi = r_replica - r_target
    tracking_error = rendimenti_attivi.std() * np.sqrt(freq)
    information_ratio = (
        (rendimenti_attivi.mean() * freq) / tracking_error
        if tracking_error != 0
        else 0.0
    )
    correlazione = np.corrcoef(r_replica, r_target)[0, 1]

    if pesi_asset is None:
        esposizione_lorda = 1.0
    elif isinstance(pesi_asset, (int, float)):
        esposizione_lorda = float(pesi_asset)
    elif isinstance(pesi_asset, pd.Series):
        esposizione_lorda = pesi_asset.abs().sum()
    else:
        esposizione_lorda = pesi_asset.abs().sum(axis=1).mean()

    return [
        f"{cagr*100:.2f}%",
        f"{volatilita_annua*100:.2f}%",
        f"{indice_sharpe:.2f}",
        f"{max_drawdown*100:.2f}%",
        f"{tracking_error*100:.2f}%",
        f"{information_ratio:.2f}",
        f"{correlazione:.4f}",
        f"{esposizione_lorda:.2f}",
    ]


etichette_metriche = [
    "1. CAGR Geometrico",
    "2. Volatilità Annualizzata",
    "3. Indice di Sharpe (Rf=0)",
    "4. Maximum Drawdown",
    "5. Tracking Error vs Target",
    "6. Information Ratio",
    "7. Correlazione con Target",
    "8. Esposizione Lorda Media",
]

dati_per_periodo = {
    ("1. Training Set (In-Sample)", "Target Index"): calcola_metriche_istituzionali(Y_train, Y_train, e_target=True),
    ("1. Training Set (In-Sample)", "OLS Statica"): calcola_metriche_istituzionali(Y_train, rendimenti_ols_train, pesi_ols_statici),
    ("1. Training Set (In-Sample)", "RFKF Regolarizzato"): calcola_metriche_istituzionali(Y_train, rendimenti_kf_train, pesi_kf_train),
    ("2. Validation Set (OOS 1)", "Target Index"): calcola_metriche_istituzionali(Y_val, Y_val, e_target=True),
    ("2. Validation Set (OOS 1)", "OLS Statica"): calcola_metriche_istituzionali(Y_val, rendimenti_ols_val, pesi_ols_statici),
    ("2. Validation Set (OOS 1)", "RFKF Regolarizzato"): calcola_metriche_istituzionali(Y_val, rendimenti_kf_val, pesi_kf_val),
    ("3. Test Set (OOS Puro 2)", "Target Index"): calcola_metriche_istituzionali(Y_test, Y_test, e_target=True),
    ("3. Test Set (OOS Puro 2)", "OLS Statica"): calcola_metriche_istituzionali(Y_test, rendimenti_ols_test, pesi_ols_statici),
    ("3. Test Set (OOS Puro 2)", "RFKF Regolarizzato"): calcola_metriche_istituzionali(Y_test, rendimenti_kf_test, pesi_kf_test),
}
tabella_master_periodi = pd.DataFrame(
    dati_per_periodo, index=etichette_metriche
)

dati_per_modello = {
    ("1. Benchmark Target", "Train"): calcola_metriche_istituzionali(
        Y_train, Y_train, e_target=True
    ),
    ("1. Benchmark Target", "Validation (OOS 1)"): (
        calcola_metriche_istituzionali(Y_val, Y_val, e_target=True)
    ),
    ("1. Benchmark Target", "Test (OOS 2)"): calcola_metriche_istituzionali(
        Y_test, Y_test, e_target=True
    ),
    ("2. OLS Statica Baseline", "Train"): calcola_metriche_istituzionali(
        Y_train, rendimenti_ols_train, pesi_ols_statici
    ),
    ("2. OLS Statica Baseline", "Validation (OOS 1)"): (
        calcola_metriche_istituzionali(
            Y_val, rendimenti_ols_val, pesi_ols_statici
        )
    ),
    ("2. OLS Statica Baseline", "Test (OOS 2)"): (
        calcola_metriche_istituzionali(
            Y_test, rendimenti_ols_test, pesi_ols_statici
        )
    ),
    ("3. RFKF Regolarizzato", "Train"): calcola_metriche_istituzionali(
        Y_train, rendimenti_kf_train, pesi_kf_train
    ),
    ("3. RFKF Regolarizzato", "Validation (OOS 1)"): (
        calcola_metriche_istituzionali(
            Y_val, rendimenti_kf_val, pesi_kf_val
        )
    ),
    ("3. RFKF Regolarizzato", "Test (OOS 2)"): calcola_metriche_istituzionali(
        Y_test, rendimenti_kf_test, pesi_kf_test
    ),
}
tabella_master_modelli = pd.DataFrame(
    dati_per_modello, index=etichette_metriche
)

print("=" * 110)
print(
    "TABELLA RIASSUNTIVA 1: CONFRONTO DI PRESA DIRETTA PER BLOCCO TEMPORALE"
)
print("=" * 110)
display(tabella_master_periodi)

print("\n" + "=" * 110)
print("TABELLA RIASSUNTIVA 2: EVOLUZIONE E STABILITÀ PER MODELLO")
print("=" * 110)
display(tabella_master_modelli)

# %%
# ==============================================================================
# MODULO 7: RAPPRESENTAZIONE GRAFICA ISTITUZIONALE A 4 PANNELLI
# ==============================================================================
ricchezza_target = (1.0 + Y).cumprod()
ricchezza_ols = (1.0 + rendimenti_ols_totali).cumprod()
ricchezza_kf = (1.0 + rendimenti_rfkf_netti).cumprod()

rend_attivi_ols = rendimenti_ols_totali - Y
rend_attivi_kf = rendimenti_rfkf_netti - Y

te_mobile_ols = rend_attivi_ols.rolling(window=52).std() * np.sqrt(52)
te_mobile_kf = rend_attivi_kf.rolling(window=52).std() * np.sqrt(52)

fig, assi = plt.subplots(4, 1, figsize=(16, 20), sharex=True)

# Pannello 1: Traiettoria di Ricchezza Cumulata
assi[0].plot(
    ricchezza_target.index,
    ricchezza_target,
    label="Indice Target (Benchmark)",
    color="darkblue",
    linewidth=2.2,
)
assi[0].plot(
    ricchezza_ols.index,
    ricchezza_ols,
    label="Baseline OLS Statica",
    color="crimson",
    linewidth=1.5,
    linestyle="--",
)
assi[0].plot(
    ricchezza_kf.index,
    ricchezza_kf,
    label="Replica RFKF Regolarizzato (Netta)",
    color="darkgreen",
    linewidth=1.8,
)
assi[0].axvline(
    Y_val.index[0],
    color="black",
    linestyle=":",
    linewidth=2,
    label="Split Val/Test (OOS)",
)
assi[0].axvline(Y_test.index[0], color="green", linestyle=":", linewidth=2)
assi[0].set_title(
    "1. Ricchezza Cumulata: Indice Target vs Strategia RFKF Regolarizzata",
    fontsize=14,
    fontweight="bold",
)
assi[0].set_ylabel("Capitale (Base 1)")
assi[0].legend(loc="upper left", frameon=True)

# Pannello 2: Profilo del Maximum Drawdown
dd_target = 1.0 - ricchezza_target / ricchezza_target.cummax()
dd_ols = 1.0 - ricchezza_ols / ricchezza_ols.cummax()
dd_kf = 1.0 - ricchezza_kf / ricchezza_kf.cummax()

assi[1].plot(
    dd_target.index,
    dd_target * 100,
    label="Drawdown Target",
    color="darkblue",
    linewidth=1.5,
)
assi[1].plot(
    dd_ols.index,
    dd_ols * 100,
    label="Drawdown OLS",
    color="crimson",
    linewidth=1.2,
    linestyle="--",
)
assi[1].plot(
    dd_kf.index,
    dd_kf * 100,
    label="Drawdown RFKF Netto",
    color="darkgreen",
    linewidth=1.5,
)
assi[1].axvline(Y_val.index[0], color="black", linestyle=":", linewidth=2)
assi[1].axvline(Y_test.index[0], color="green", linestyle=":", linewidth=2)
assi[1].set_title(
    "2. Profilo di Drawdown Peak-to-Valley (%)",
    fontsize=14,
    fontweight="bold",
)
assi[1].set_ylabel("Drawdown (%)")
assi[1].legend(loc="upper left", frameon=True)

# Pannello 3: Rendimenti Attivi (Residui di Replica e_t)
assi[2].plot(
    rend_attivi_kf.index,
    rend_attivi_kf * 100,
    color="purple",
    linewidth=1.0,
    alpha=0.75,
    label="Active Return Settimanale RFKF ($e_t$ netto)",
)
assi[2].axhline(0, color="black", linestyle="-", linewidth=1.2)
assi[2].axvline(Y_val.index[0], color="black", linestyle=":", linewidth=2)
assi[2].axvline(Y_test.index[0], color="green", linestyle=":", linewidth=2)
assi[2].set_title(
    "3. Rendimenti Attivi Settimanali (Residuo di Replica: RFKF - Target)",
    fontsize=14,
    fontweight="bold",
)
assi[2].set_ylabel("Active Return (%)")
assi[2].legend(loc="upper left", frameon=True)

# Pannello 4: Tracking Error Mobile Annualizzato (Finestra 52 Settimane)
assi[3].plot(
    te_mobile_ols.index,
    te_mobile_ols * 100,
    label="Rolling TE 1 Anno - OLS Statica",
    color="crimson",
    linewidth=1.5,
    linestyle="--",
)
assi[3].plot(
    te_mobile_kf.index,
    te_mobile_kf * 100,
    label="Rolling TE 1 Anno - RFKF Regolarizzato (Netto)",
    color="darkred",
    linewidth=2.0,
)
assi[3].axvline(Y_val.index[0], color="black", linestyle=":", linewidth=2)
assi[3].axvline(Y_test.index[0], color="green", linestyle=":", linewidth=2)
assi[3].set_title(
    "4. Tracking Error Mobile Annualizzato (Finestra Rolling 52 Settimane)",
    fontsize=14,
    fontweight="bold",
)
assi[3].set_xlabel("Data di Rilevazione", fontsize=12)
assi[3].set_ylabel("Tracking Error (%)")
assi[3].legend(loc="upper left", frameon=True)

plt.tight_layout()
plt.show()
```
